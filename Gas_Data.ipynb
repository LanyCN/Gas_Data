```python
#criando uma cópia para analisar esses valores com tipos errados
data_pre = data.copy()
```

```python
data_pre['DATA INICIAL'] = pd.to_datetime(data_pre['DATA INICIAL'])
data_pre['DATA FINAL'] = pd.to_datetime(data_pre['DATA FINAL'])
```

```python
#verificado quais colunas deveriam ter valores numéricos
data_pre.info()
```

```python
#convertendo valores para numeric, usando a funcão pd.numeric
#de acordo com a documentação do Pandas, das três opções de saída para essa conversão (ignore, raise or coerce), e que por default é raise, ficarei com a opção coerce para definir como nulos os valores que não correspondem a números
for atributo in ['MARGEM MÉDIA REVENDA', 'PREÇO MÉDIO DISTRIBUIÇÃO', 'DESVIO PADRÃO DISTRIBUIÇÃO', 'PREÇO MÍNIMO DISTRIBUIÇÃO', 'PREÇO MÁXIMO DISTRIBUIÇÃO', 'COEF DE VARIAÇÃO DISTRIBUIÇÃO']:
    data_pre[atributo] = pd.to_numeric(data_pre[atributo], errors='coerce')
```
